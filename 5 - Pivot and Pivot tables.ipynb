```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Brand': ['Tata', 'Tesla', 'Audi', 'BMW'],
    'Location': ['India', 'US', 'Germany', 'Germany'],
    'Model':['Ace', '3', 'Q3', 'Z4'],
    'Sales': [100,75,84,92]
})
df
```

```python
pd.pivot(data=df, index='Brand', columns='Location', values=['Model', 'Sales'])
```

```python
tesla_s = pd.Series(['Tesla', 'US', 'S', 40], index=['Brand', 'Location', 'Model', 'Sales'])
tesla_s
```

```python
df = df.append(tesla_s, ignore_index=True)
df
```

```python
pd.pivot_table(data=df, index='Brand', columns='Location', values=['Sales'], aggfunc='min')
```

```python
b = pd.read_csv('data/AirBnb_NYC_2019.csv', nrows=10000)
b.head()
```

```python
pd.pivot_table(index=['neighbourhood_group', 'neighbourhood'], data=b, columns=['room_type'],
               aggfunc='mean', values=['price', 'minimum_nights'])
```
